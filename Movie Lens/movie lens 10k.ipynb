## 각 상품별로(row)별로 row(user_id), column(movie), rating 데이터를 recommend matrix에 기록해라

```r
recommend.matrix <- data.frame()
```

```r
str(recommend.matrix)
```

```r
for(i in 1:recommend.raw.matrix.row){
    mat = data.frame("movie_id"=rep(i,recommend.raw.matrix.row),"recommend_id"=1:recommend.raw.matrix.row,
           "score"=t(t(recommend.raw.matrix[i,])))
    recommend.matrix <- rbind(recommend.matrix,mat);
}
```

```r
head(recommend.matrix)
```

```r
recommend.matrix.df <- tbl_df(recommend.matrix)

find_recommend <- function(id, n){
  res <- select(arrange(filter(recommend.matrix.df,movie_id == id,
                        recommend_id != id),desc(score)),movie_id,recommend_id,score)[1:n,]
  return(as.data.frame(res))
}
```

```r
select(arrange(filter(recommend.matrix.df,movie_id == 1,
                        recommend_id != 1),desc(score)),movie_id,recommend_id,score)[1:30,]
```

```r
find_recommend(1,30)
```

```r
colnames(movie_title) <- c("movie_id","title")
```

```r
find_recommend2 <- function(id, n){
  res <- select(arrange(filter(recommend.matrix.df,movie_id == id,
                        recommend_id != id),desc(score)),recommend_id,score)[1:n,]
  res.df <- as.data.frame(res)
  colnames(res.df) <- c("movie_id","score")  
  res.df <- merge(res.df,movie_title,by="movie_id")  
  return(res.df)
}
```
